```python
# ONLY RUN ONCE
# As Incidence rate is directly link with the target we cannot use it on the test set,
# so we have to create a list of dataset to merge it on the test set
list_to_datasets_Incidence=list()
for Categorical_Variable in Cat:
    print(str(Categorical_Variable))
#     Calculation of the incidence rate for each Categorical Variable  
    x=(train[[Categorical_Variable,'HasDetections']].groupby(Categorical_Variable).count()/len(train)).reset_index()
#     Renaming the column "Incidence" + "name of the original variable"
    x=x.rename({'HasDetections':str("Incidence_"+str(Categorical_Variable))},axis=1)
#     appending incidence rate datasets per variable to the list 
    list_to_datasets_Incidence.append(x)
#     Adding the column "Incidence" + "name of the original variable" to the train set 
    train=pd.merge(train,x,on=Categorical_Variable,how='left')
    train=train.drop(Categorical_Variable,axis=1)
```

```python
# ONLY RUN ONCE
# Merge incidence rate calculated in the Train dataset to the test dataset
for datasets in list_to_datasets_Incidence:
    print(datasets.columns[0])
    test=pd.merge(test,datasets,on=datasets.columns[0],how='left')
```
