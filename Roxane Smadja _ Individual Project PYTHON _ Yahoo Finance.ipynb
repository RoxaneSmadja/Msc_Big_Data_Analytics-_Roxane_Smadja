```python
#Function to get my 5 stocks in one table 

def myFinFunction(*args, start_date, end_date):
    
    stocks_prices = pd.DataFrame()
    
    for ticker in args:
        temp = get_data(ticker , start_date=start_date , end_date=end_date)
        stocks_prices = stocks_prices.append(temp)
        
    return stocks_prices[['ticker', 'open', 'close', 'volume']]

stocks_prices = myFinFunction('MSFT', 'ADBE', 'NLOK', 'DELL', 'CRM', start_date='15/02/2020' , end_date='15/04/2020')
stocks_prices
```

```python
#Get the balance sheets of my five companies in one dataframe
tickers=['MSFT', 'ADBE', 'DELL', 'NLOK', 'CRM']

balance_sheets = {}
for ticker in tickers:
    balance_sheets[ticker] = get_balance_sheet(ticker)

recent_sheets = {ticker : sheet.iloc[:,:1] for ticker,sheet in balance_sheets.items()}
recent_sheets = {ticker : sheet.iloc[:,:1] for ticker,sheet in balance_sheets.items()}
 
for ticker in recent_sheets.keys():
    recent_sheets[ticker].columns = ["Recent"]

combined_sheets = pd.concat(recent_sheets)

combined_sheets = combined_sheets.reset_index()


combined_sheets.columns = ["Ticker", "Fin_Aggregate", "Recent"]
combined_sheets= combined_sheets.pivot(index='Ticker', columns='Fin_Aggregate', values='Recent')
combined_sheets
```

```python
#Get the cash flow statements of my five companies in one dataframe

tickers=['MSFT', 'ADBE', 'DELL', 'NLOK', 'CRM']

cash_flows = {}
for ticker in tickers:
    cash_flows[ticker] = get_cash_flow(ticker)

recent_cash_flows = {ticker : flow.iloc[:,:1] for ticker,flow in cash_flows.items()}
 
 
for ticker in recent_cash_flows.keys():
    recent_cash_flows[ticker].columns = ["Recent"]
 
 
combined_cash_flows = pd.concat(recent_cash_flows)
 
combined_cash_flows = combined_cash_flows.reset_index()
 
combined_cash_flows.columns = ["Ticker", "Fin_Aggregate", "Recent"]

combined_cash_flows= combined_cash_flows.pivot(index='Ticker', columns='Fin_Aggregate', values='Recent')
combined_cash_flows.dropna
combined_cash_flows
```

```python
#Get the income statements of my five companies in one dataframe
tickers=['MSFT', 'ADBE', 'DELL', 'NLOK', 'CRM']

income_statements = {}
for ticker in tickers:
    income_statements[ticker] = get_income_statement(ticker)
    
    recent_income_statements = {ticker : sheet.iloc[:,:1] for ticker,sheet in income_statements.items()}
 
for ticker in recent_income_statements.keys():
    recent_income_statements[ticker].columns = ["Recent"]
 
combined_income = pd.concat(recent_income_statements)
 
combined_income = combined_income.reset_index()
 
combined_income.columns = ["Ticker", "Fin_Aggregate", "Recent"]

combined_income=combined_income.pivot(index='Ticker', columns='Fin_Aggregate', values='Recent')
```

```python
#Merging of the income statement and the balance sheets in order to add some ratios below
FIN_STATEMENTS=pd.merge(combined_income, combined_sheets, how='right', on='Ticker')
```

```python
#Calculation of the financial ratios
FIN_STATEMENTS['Debtor_days']=(FIN_STATEMENTS['netReceivables']/FIN_STATEMENTS['totalRevenue'])*365
FIN_STATEMENTS['Days_payable_outstanding']=(FIN_STATEMENTS['accountsPayable']/FIN_STATEMENTS['costOfRevenue'])*365
FIN_STATEMENTS['Asset_turnover_ratio']=FIN_STATEMENTS['totalRevenue']/FIN_STATEMENTS['totalAssets']
FIN_STATEMENTS['Days_in_inventory']=(FIN_STATEMENTS['inventory']/FIN_STATEMENTS['costOfRevenue'])*365
FIN_STATEMENTS['Gross_Profit_margin'] = FIN_STATEMENTS['grossProfit']/FIN_STATEMENTS['totalRevenue']

```

```python
import pandas as pd
import numpy as np

def wgs84_to_web_mercator(df, lon="lon", lat="lat"):
    """Converts decimal longitude/latitude to Web Mercator format"""
    k = 6378137
    df["x"] = df[lon] * (k * np.pi/180.0)
    df["y"] = np.log(np.tan((90 + df[lat]) * np.pi/360.0)) * k
    return df

df = pd.DataFrame(dict(City=["Redmond", "San Jose", "Round Rock", "Tempe", "San Francisco"  ], lon=[-97.478256, -121.89496, -97.678894, -111.939896,-122.431297], lat=[35.657295, 37.33939, 30.508255, 33.427204, 37.773972], Company=['MSFT', 'ADBE', 'DELL', 'NLOK', 'CRM']))
wgs84_to_web_mercator(df)

```
